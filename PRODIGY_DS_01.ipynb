Part 1: Import Libraries and Load Data


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the total population dataset, skipping the first 4 rows
Population = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_1052975.csv", skiprows=4)

# Display the first few rows of the dataset to check its structure and content
Population.head()
```

Part 2: Inspect and Clean Total Population Data


```python
# Display a summary of the DataFrame, including the number of non-null entries, data types of each column, and memory usage
Population.info()
```

```python
# Generate descriptive statistics for the DataFrame, including count, mean, standard deviation, min, max, and quartiles for numerical columns
Population.describe()
```

```python
# Print the column names of the DataFrame to see the structure
print(Population.columns)
# Drop the 'Indicator Name' and 'Country Code' columns from the DataFrame as they are not needed for further analysis
# The 'inplace=True' argument modifies the DataFrame directly without needing to assign the result to a new variable
Population.drop(columns=['Indicator Name', 'Country Code'], inplace=True)
```

```python
# Check if there are any null values in the DataFrame
print(Population.isnull().values.any())
# Remove all rows that contain any null values from the DataFrame
Population_cleaned = Population.dropna()
# Check again if there are any null values in the cleaned DataFrame to confirm that all null values have been removed
print(Population_cleaned.isnull().values.any())
```

Part 3: Load and Clean Male Population Data


```python
# Load the male population dataset, skipping the first 4 rows
PopulationMale = pd.read_csv("API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_1035150.csv", skiprows=4)
# Drop the 'Indicator Name', 'Country Code', and 'Unnamed: 68' columns from the DataFrame as they are not needed for further analysis
# The 'Unnamed: 68' column is removed as it is likely irrelevant or empty
PopulationMale.drop(columns=['Indicator Name', 'Country Code','Unnamed: 68'], inplace=True)
# Check if there are any null values in the DataFrame
print(PopulationMale.isnull().values.any())
# Remove all rows that contain any null values from the DataFrame
Population_cleanedMale = PopulationMale.dropna()
# Check again if there are any null values in the cleaned DataFrame to confirm that all null values have been removed
print(Population_cleanedMale.isnull().values.any())
```

Part 4: Load and Clean Female Population Data


```python
# Load the female population dataset, skipping the first 4 rows
PopulationFemale = pd.read_csv("API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1292964.csv", skiprows=4)
# Drop the 'Indicator Name', 'Country Code', and 'Unnamed: 68' columns from the DataFrame as they are not needed for further analysis
# The 'Unnamed: 68' column is removed as it is likely irrelevant or empty
PopulationFemale.drop(columns=['Indicator Name', 'Country Code','Unnamed: 68'], inplace=True)
# Check if there are any null values in the DataFrame
print(PopulationFemale.isnull().values.any())
# Remove all rows that contain any null values from the DataFrame
Population_cleanedFemale = PopulationFemale.dropna()
# Check again if there are any null values in the cleaned DataFrame to confirm that all null values have been removed
print(Population_cleanedFemale.isnull().values.any())
```

Part 5: Merge Data and Calculate Gender Distribution for Selected Countries



```python
# Extract the relevant data for the latest year (2023)
latest_year = '2023'
total_pop = Population_cleaned[['Country Name', latest_year]].rename(columns={latest_year: 'Total'})
male_pop = Population_cleanedMale[['Country Name', latest_year]].rename(columns={latest_year: 'Male'})
female_pop = Population_cleanedFemale[['Country Name', latest_year]].rename(columns={latest_year: 'Female'})
# Merge the dataframes on 'Country Name'
population_merged = pd.merge(total_pop, male_pop, on='Country Name')
population_merged = pd.merge(population_merged, female_pop, on='Country Name')

# List of countries to visualize
selected_countries = ['United Arab Emirates', 'Qatar', 'Latvia', 'Ukraine', 'Rwanda', 'Nepal', 'China', 'India', 'Portugal', 'Japan']

# Filter the dataframe to include only the selected countries
population_selected = population_merged[population_merged['Country Name'].isin(selected_countries)].copy()

# Calculate gender distribution (as percentages of the total population)
population_selected.loc[:, 'Male (%)'] = (population_selected['Male'] / population_selected['Total']) * 100
population_selected.loc[:, 'Female (%)'] = (population_selected['Female'] / population_selected['Total']) * 100

# Display the selected countries dataframe (optional)
print(population_selected)
```

Part 6: Create and Display Bar Chart


```python
# Create a bar chart
fig, ax = plt.subplots(figsize=(12, 8))  # Adjust the figure size

bar_width = 0.4
index = population_selected['Country Name']

bar1 = ax.bar(index, population_selected['Male (%)'], bar_width, label='Male (%)', color='b')
bar2 = ax.bar(index, population_selected['Female (%)'], bar_width, bottom=population_selected['Male (%)'], label='Female (%)', color='r')

# Add labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Percentage')
ax.set_title('Gender Distribution in Selected Countries')
ax.legend()

# Add data labels on the bars
for bar in bar1 + bar2:
    height = bar.get_height()
    ax.annotate(f'{height:.1f}%', xy=(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='center', fontsize=9, color='white')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a grid
ax.yaxis.grid(True)

# Show the bar chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define column names based on the provided data format
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                 'marital-status', 'occupation', 'relationship', 'race', 'sex',
                 'capital-gain','capital-loss',
                 'hours-per-week', 'native-country', 'income']

# Load the dataset
data = pd.read_csv('adult.data', header=None, names=column_names, na_values=' ?', skipinitialspace=True)

# Filter for people from the United States
data_us = data[data['native-country'] == 'United-States']

print(data_us['age'].describe())


data_us = data_us.dropna(subset=['age'])

# Create the histogram
plt.figure(figsize=(12, 8))
plt.hist(data_us['age'], bins=range(0, 95, 5), edgecolor='black', alpha=0.7)  # Bins of 5 years
plt.title('Distribution of Ages in the United States Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.xticks(range(0, 95, 5))  # Set x-ticks to match bins
plt.grid(True)
plt.show()
```
